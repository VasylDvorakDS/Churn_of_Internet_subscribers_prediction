### Создание когорт (сегментов) по колонке Tenure

**Для начала поместим различные значения tenure (1 месяц, 2 месяца, 3 месяца и так далее) в отдельные когорты (сегменты).**

**Для каждого уникального значения колонки tenure в качестве когорты, вычислим процент оттока (churn rate) - это количество тех людей, кто ушёл в отток в когорте, как процент от общего количества людей в данной когорте. Вычислим такой процент отдельно для каждой когорты. Например, для когорты с tenure = "1 месяц" должен получиться процент оттока 61.99%. В итоге у нас должны получиться когорты от 1 до 72 месяцев, и с увеличением количества месяцев уровень оттока должен снижаться. Это разумно, потому что чем дольше человек пользуется услугами компании, тем вероятнее то, что его/её всё устраивает, и он/она продолжит пользоваться этими услугами.**

```python
churn_yes=df.groupby(['Churn','tenure']).count().transpose()['Yes']
churn_no=df.groupby(['Churn','tenure']).count().transpose()['No']
churn_rate=100*churn_yes/(churn_no+churn_yes)
churn_ratepercent=churn_rate.transpose()['customerID']
churn_ratepercent
```

**Используя результаты предыдущего задания, построим график процента оттока для различных значений tenure.**

```python
plt.figure(figsize=(10,6),dpi=200)
churn_rate.transpose()['customerID'].plot()
plt.ylabel('Churn Percentage');
```

### Более крупные когорты (сегменты)
**На основе колонки tenure создадим новую колонку с названием "Tenure Cohort", в которой будут следующие 4 категории:**
   * '0-12 месяцев'
   * '24-48 месяцев'
   * '12-24 месяцев'
   * 'Более 48 месяцев'    

```python
def monthes_seporation(monthes):
        if 0<monthes<13:
            return '0-12 месяцев'
        elif 13<=monthes<25:
            return '12-24 месяцев'
        elif 25<=monthes<49:
            return '24-48 месяцев'
        else:
            return'Более 48 месяцев' 
```

```python
df["Tenure Cohort"]=df['tenure'].apply(lambda monthes: monthes_seporation(monthes))
df["Tenure Cohort"]
```

**Построим график scatterplot для Total Charges и Monthly Charts, раскрашивая график разными цветами по колонке "Tenure Cohort", которую Мы создали в предыдущем задании.**

```python
plt.figure(figsize=(10,3),dpi=200)
sns.scatterplot(x='MonthlyCharges',y='TotalCharges',data=df,hue='Tenure Cohort',alpha=0.5,palette='Dark2');
```

**Построим график countplot с количеством ушедших и не ушедших в отток людей в каждой когорте.**

# Часть 4: Предиктивная модель

**Исследуем 4 модели на основе деревьев: одно дерево решений, случайный лес, адаптивный бустинг и градиентный бустинг.**


## Одно дерево решений (Decision Tree)

**Разделиим все данные на признаки X и целевую переменную Y. Создадим дамми-переменные, где это необходимо, а также посмотрим, есть ли какие-то признаки, которые можно было бы удалить.**

```python
X = df.drop(['Churn', 'customerID'],axis=1)
X = pd.get_dummies(X,drop_first=True)
y = df['Churn']
```

**Выполним разбиение данных на обучающий и тестовый наборы (train test split), откладывая в сторону 10% данных для тестирования. В решениях мы используем random_state=101.**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

**Построим модель Decision Tree и оценим её работу. Для этого выполним следующие шаги:**
   1. Обучим модель дерева решений (выполним поиск по сетке для поиска оптимальных значений гиперпараметров).
   2. Оценим метрики работы модели дерева решений - для этого выведим отчёт "classification report" и нарисуем график с матрицей ошибок (confusion matrix).
   2. Посмотрим на важность признаков для получившегося дерева решений.
   4. Нарисуем график получившегося дерева. Обратим внимание, что если дерево получилось очень большим и мы не выполняли усечение дерева с помощью параметров, то вызов метода plot_tree может привести к ошибке.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
from sklearn.tree import plot_tree
```

```python
param_grid = {'max_leaf_nodes':[20,50],'max_depth':[3,6,8], 'criterion':['gini','entropy']}
```

```python
model_originnal = DecisionTreeClassifier()
grid_model=GridSearchCV(model_originnal, param_grid)
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
best_params=grid_model.best_params_
best_params
```

```python
model = DecisionTreeClassifier(max_depth=best_params['max_depth'], max_leaf_nodes=best_params['max_leaf_nodes'], 
                               criterion=best_params['criterion'])
```

```python
model.fit(X_train,y_train)
```

```python

model_pred = model.predict(X_test)
confusion_matrix(y_test,model_pred)
```

```python
plot_confusion_matrix(model,X_test,y_test);
```

```python
print(classification_report(y_test,model_pred))
```

```python
imp_feats = pd.DataFrame(data=model.feature_importances_,index=X.columns,
                         columns=['Feature Importance']).sort_values("Feature Importance")
minimum=imp_feats['Feature Importance'].max()/100
imp_feats=imp_feats.loc[imp_feats['Feature Importance'] > minimum]
imp_feats
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.barplot(data=imp_feats,x=imp_feats.index,
            y='Feature Importance')
plt.xticks(rotation=90)
plt.title("Feature Importance for Decision Tree");
```

```python

    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X.columns);
```

## Случайный лес (Random Forest)

*Создадим модель случайного леса, затем выведим отчёт "classification report" и матрицу ошибок для предсказанных результатов на тестовом наборе данных.**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_originnal = RandomForestClassifier()
param_grid = {'n_estimators':[3,5,6,7],'max_depth':[3,6,8], 'criterion':['gini','entropy'], 
              'max_features' : ["auto", "sqrt", "log2"]}
grid_model=GridSearchCV(model_originnal, param_grid)
grid_model.fit(X_train,y_train)
grid_model.best_estimator_
```

```python
best_params=grid_model.best_params_
best_params
```

```python
model = RandomForestClassifier(max_depth=best_params['max_depth'], max_features=best_params['max_features'], 
                               criterion=best_params['criterion'], n_estimators=best_params['n_estimators'])
model.fit(X_train,y_train)
model_pred = model.predict(X_test)
confusion_matrix(y_test,model_pred)
```

```python
plot_confusion_matrix(model,X_test,y_test);
```

```python
print(classification_report(y_test,model_pred))
```

## Расширяемые деревья (Boosted Trees)

**Создадим модель AdaBoost или Gradient Boosting, затем выведим отчёт "classification report" и матрицу ошибок для предсказанных результатов на тестовом наборе данных.**

```python
from sklearn.ensemble import AdaBoostClassifier
model_originnal = AdaBoostClassifier()
param_grid = {'n_estimators':[3,5,6,7], 'algorithm' : ['SAMME', 'SAMME.R']}

```

```python
grid_model=GridSearchCV(model_originnal, param_grid)
grid_model.fit(X_train,y_train)
grid_model.best_estimator_
```

```python
best_params=grid_model.best_params_
best_params
```

```python
model = AdaBoostClassifier(algorithm=best_params['algorithm'], n_estimators=best_params['n_estimators'])
model.fit(X_train,y_train)
model_pred = model.predict(X_test)
confusion_matrix(y_test,model_pred)
```

```python
plot_confusion_matrix(model,X_test,y_test);
```

```python
print(classification_report(y_test,model_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_originnal = GradientBoostingClassifier()
param_grid = {'n_estimators':[3,5,6,7], 'max_depth' : [3, 5,6]}
grid_model=GridSearchCV(model_originnal, param_grid)
grid_model.fit(X_train,y_train)
grid_model.best_estimator_
```

```python
model = GradientBoostingClassifier(max_depth=best_params['max_depth'], n_estimators=best_params['n_estimators'])
model.fit(X_train,y_train)
model_pred = model.predict(X_test)
confusion_matrix(y_test,model_pred)
```

```python
plot_confusion_matrix(model,X_test,y_test);
```

```python
print(classification_report(y_test,model_pred))
```
